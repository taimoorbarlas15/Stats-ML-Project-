## Z-score

Detect Outliers using Z-Score. (Set threshold =3)

```python
import pandas as pd
from scipy.stats import zscore
```

```python
z_scores =  numeric_columns.apply(zscore)
```

```python
threshold = 3
```

```python
#Idenctify ouliers in the dataset
```

```python
outliers = (z_scores.abs() > threshold).any(axis=1)
```

```python
# Display outliers
```

```python
outliers_data = numeric_columns[outliers]
```

```python
print(outliers_data)
```

## Remove outliers

```python
from scipy.stats import zscore
```

```python
# Remove outliers
```

```python
# Recalculate Z-scores with NumPy imported
```

```python
z_scores = zscore(numeric_columns.fillna(numeric_columns.mean()), nan_policy='omit')
```

```python
abs_z_scores = np.abs(z_scores)
```

```python
# Filter data: keep only rows where all Z-scores are below the threshold of 3
```

```python
filtered_data = numeric_columns[(abs_z_scores < 3).all(axis=1)]
```

```python
print(filtered_data)
```

### Standardization and Normalization

Apply standardization and normalization on the columns

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

```

## Standardization

```python
scaler_standard = StandardScaler()
```

```python
scaler_minmax = MinMaxScaler()
```

```python
# Apply standardization
```

```python
standardized_data = scaler_standard.fit_transform(filtered_data.fillna(filtered_data.mean()))
```

```python
standardized_df = pd.DataFrame(standardized_data, columns=filtered_data.columns)
```

```python
# Display standardized and normalized dataframes
```

```python
standardized_df.head()
```

## Normalization

```python
#Apply normalization 
```

```python
normalized_data = scaler_minmax.fit_transform(filtered_data.fillna(filtered_data.mean()))
```

```python
normalized_df = pd.DataFrame(normalized_data, columns=filtered_data.columns)
```

```python
# Display  normalized dataframes
```

```python
normalized_df.head()
```

## Distribution check

```python
# Set up the figure and axes for the histograms
```

```python
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 10))
fig.suptitle('Distributions of Standardized and Normalized Data', fontsize=16)

# Plot histograms for standardized data
for idx, column in enumerate(standardized_df.columns):
    ax = axes[0, idx % 4]
    standardized_df[column].hist(ax=ax, bins=20, alpha=0.7)
    ax.set_title(f'Std: {column}')

    if idx == 3:  # Break after 4 plots for the first row
        break

# Plot histograms for normalized data
for idx, column in enumerate(normalized_df.columns):
    ax = axes[1, idx % 4]
    normalized_df[column].hist(ax=ax, bins=20, alpha=0.7, color='orange')
    ax.set_title(f'Norm: {column}')

    if idx == 3:  # Break after 4 plots for the first row
        break

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
